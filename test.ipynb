```python
from demoparser2 import DemoParser

# Initialize the parser
parser = DemoParser("test_demo.dem")
wanted_ticks = parser.parse_event('round_poststart')['tick'].tolist()

df = parser.parse_ticks(['team_name',
                         'total_rounds_played',
                         'inventory',
                         'armor_value',
                         'cash_spent_this_round',
                         'round_start_equip_value',
                         'current_equip_value',
                         'balance',
                         'has_helmet'],
                        ticks=wanted_ticks)

sorted_df = df.sort_values(by=['tick', 'team_name'], ascending=[True, False])
sorted_df.head(10)
```
